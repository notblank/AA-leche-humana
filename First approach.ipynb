# AA selection 

```r
aaSelect <- function(df, AA){
    aaDf <- df %>% select(id = ID, group = GROUP, age = MATERNALAGE, sex = SEXO, starts_with(AA)) %>%
                         mutate(id = factor(id), 
                                sex =  factor(sex, labels = c('girl', 'boy')),
                                group = factor(group, labels = c('teen', 'adult'))) 
    
    return(aaDf)
}
```

```r
aaLong <-  function(df, AA){
    aaL <- df %>% gather(week, level, starts_with(AA))
    
    aaL[aaL == paste(AA, 'Calostrum', sep = '')] = 1
    aaL[aaL == paste(AA, 'Transition', sep = '')] = 2
    aaL[aaL == paste(AA, 'Mature2m', sep = '')] = 8
    aaL[aaL == paste(AA, 'Mature4m', sep = '')] = 16

    aaL <- aaL %>% mutate(week = as.numeric(week))
    
    return(aaL)
}
```

```r
AAdfById <- AAdfLongComplete %>% group_by(id, sex, group) %>% nest()
head(AAdfById)
```

```r
sexModels <- function(df){
    
    mS <- lm(level~week, data = df)
    
    return(mS)
}
```

```r
AAdfById <- AAdfById %>% mutate(
                            model = map(data, sexModels)
                                )
head(AAdfById)
```

```r
AAdfById$model[[1]]
```

```r
AAdfById <- 
    AAdfById %>% mutate(
        glance = model %>% map(broom::glance),
        tidy = model %>% map(broom::tidy)#,
        #intercept = tidy %>% map_dbl("estimate")
    )

AAdfById$glance[[1]]
AAdfById$tidy[[1]]
```

```r
rdf <- AAdfById %>% unnest(glance) %>% select(r.squared) %>% filter(r.squared != 0)

ggplot(rdf, aes(x ='', y = r.squared)) + geom_boxplot() + geom_point(alpha = 0.3)
```
